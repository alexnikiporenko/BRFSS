## 1. Relationship between the intakes of healthy foods

Cross-selling is a strategy of advertising products that are related to the products currently bought by the customers [1, p. 164]. In the context of sales of plant-based foods, grocery stores would benefit from the knowledge of whether a customer purchasing one type of these food items would be more likely to purchase another. Furthermore, this data would be of interest to public health researchers studying “food deserts”, areas where healthy food is inaccessible [2]. This paper will seek an answer to the following question:

**Does regularly including a type of plant-based food in a person’s diet make them more likely to include other plant-based foods too?**

```python
# loading libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import seaborn as sns
```

```python
# encoding is specified, as Python throws a UnicodeDecodeError:
# 'utf-8' codec can't decode byte 0x92 in position 3: invalid start byte.
# low_memory=False is set as columns have mixed types
df = pd.read_csv("NYSDOH_BRFSS_SurveyData_2015.csv", encoding='cp1252', low_memory=False)

# viewing the first few rows of the dataset
df.head()
```

To choose a suitable technique for each question, I started by looking at the data types of the attributes relevant to my queries.

For my first research question, I have selected six attributes representing intakes of various plant-based foods. The values are integers, with two implied decimal places [3], and so I have decided to use linear regression, often used on numeric attributes [1, p. 114]. To measure the relationship between the food-intakes, I am going to use the Pearson correlation that represents the degree to which the chosen pairs of attributes have a linear relationship. Furthermore, I will then model the effect of multiple food-type attributes on a single attribute using multiple linear regression, a technique for mapping multiple numeric inputs to a single output [1, p. 120]. 

```python
# choosing 6 attributes
diet_df = df[["FTJUDA1_", "FRUTDA1_", "BEANDAY_", "GRENDAY_", "ORNGDAY_", "VEGEDA1_"]]

# dropping missing values
diet_df = diet_df.dropna().reset_index(drop=True)

# renaming columns for clarity
diet_df.columns = ['FRUIT_JUICE', 'FRUIT', 'BEANS', 'GREEN_VEG', 'ORANGE_VEG', 'OTHER_VEG']

# viewing the first few rows of the dataset
diet_df.head()
```

First, I have calculated the Pearson correlation coefficients between the intakes of plant-based foods:

```python
# visualising a correlation matrix using seaborn
corr_matrix = diet_df.corr()
mask = np.zeros_like(corr_matrix, dtype=np.bool)
mask[np.triu_indices_from(mask)]= True
f, ax = plt.subplots(figsize=(10, 15))

heatmap = sns.heatmap(corr_matrix,
                      mask = mask,
                      square = True,
                      linewidths = .5,
                      cmap = 'Greens',
                      cbar_kws = {'shrink': .4,
                                'ticks' : [0, 0.1, 0.2, 0.3, 0.4]},
                      vmin = 0,
                      vmax = 0.4,
                      annot = True,
                      annot_kws = {'size': 12})

#add the column names as labels
ax.set_yticklabels(corr_matrix.columns, rotation = 0)
ax.set_xticklabels(corr_matrix.columns)

sns.set_style('whitegrid')
```

Then, I have built six multiple linear regression models to predict the intake of each food-type using other produce as an input, with the following results:

```python
# creating a dictionary to fill in with "food-type":"r2 score" key-value pairs
mult_lin_regressions = {}

for i in range(6):
    columns = diet_df.columns.tolist()
    column = columns[i]
    columns.remove(column)
    X = diet_df[columns]
    y = diet_df[column]
    reg = LinearRegression().fit(X, y)
    mult_lin_regressions[column] = round(reg.score(X, y), 4)

# viewing the scores
print(mult_lin_regressions)
```

```python
# function to add value labels
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i]/2,y[i], ha = 'center', bbox = dict(facecolor = 'white', alpha = .5))
        
# creating data for the bar chart
x = list(mult_lin_regressions.keys())
y = list(mult_lin_regressions.values())

# configuring plot's size and grid
plt.figure(figsize=(10, 6))
plt.grid(b=None)

# making the bar chart on the data
plt.bar(x, y, color="green")
      
# calling the function to add value labels
addlabels(x, y)
    
# giving X and Y labels
plt.xlabel("Food type")
plt.ylabel("Coefficient of determination ($R^2$)")

# visualizing the plot
plt.show()
```

The regression models built for the first question indicate that out of the food-types present in the survey, green vegetables have the strongest correlation with other produce. Pairing “GREEN_VEG” with any of the other food-type attributes (except for fruit juice, which appears to have almost no relationship to other foods), we get a Pearson coefficient of $r ≈ ±0.3$, indicating a weak linear relationship [1, p. 106]. Furthermore, if we predict the intake of green vegetables using a multiple linear regression function that uses all other food-type attributes as inputs, we get an $R^2$ value of 0.255. This means that people who regularly include green vegetables in their diet are indeed more likely to include other fruits and vegetables too. Poverty is one way to explain this disparity – the price-per-calorie of energy-dense junk foods is a tenth of healthier foods [4].

## 2. Effects of past smoking in the old age

Tobacco smoking is the leading cause of preventable death and disease [5]. The awareness of this has led to various tobacco control strategies and the subsequent increase in the number of quitting smokers [6]. When viewed in the context of the USA’s ageing population [7], the fact that smoking was more prevalent in the past means that there is a sizeable proportion of ex-smokers amongst the current non-smokers in the senior age group. The older population’s greater healthcare needs [8] require healthcare providers to have a thorough understanding of the causes of ill health. With that in mind, this paper asks:

**Does past smoking have an effect on health in old age?**

```python
# loading libraries
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn import metrics
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
```

```python
# selecting attributes
smoking_seniors_df = df[["GENHLTH", #health indicators - use all to predict smoker status. 
                        "PHYSHLTH",
                        "MENTHLTH",
                        "BPHIGH4",
                        "BLOODCHO",
                        "CVDINFR4",
                        "CVDCRHD4",
                        "CVDSTRK3",
                        "ASTHMA3",
                        "CHCSCNCR",
                        "CHCOCNCR",
                        "CHCCOPD1",
                        "HAVARTH3",
                        "ADDEPEV2",
                        "CHCKIDNY",
                        "DIABETE3",
                        "QLACTLM2",
                        "USEEQUIP",
                        "BLIND",
                        "DECIDE",
                        "DIFFWALK",
                        "DIFFDRES",
                        "DIFFALON",
                        "EXERANY2",
                        "_BMI5CAT",
                        "_IMPAGE", # to separate old age category
                        "_SMOKER3" # smoker status
                        ]]

# viewing the first few rows of the dataset
smoking_seniors_df.head()
```

Dealing with missing values is an important step in data exploration [1, p. 61]. I have used Pandas unique() function to examine the values in each column, and noticed that the dataset uses numerous ways to represent missing values: “Not asked or Missing”, “Refused”, “Don't know/Not sure” and “Don't know/Refused/Missing”. Furthermore, I have spotted a few inconsistencies – in some cases, “Don't know/Not sure” has spaces on either side of the slash, and also occasionally the apostrophe in “Don't” is written as an acute accent (“Don´t”). I have created a list of these missing values (including inconsistent spellings) and reimported the dataset, assigning my list to “na_values” – a parameter enabling Pandas to recognise additional strings as NA/NaN [9, p. 172].

```python
# viewing unique values in each column to identify missing values
for col in smoking_seniors_df:
    print(smoking_seniors_df[col].unique())
```

```python
# Replacing missing values with NaNs.
# Confusingly, the dataset contains numerous different symbols instead of an apostrophe 
smoking_seniors_df = smoking_seniors_df.replace(["Don’t know/Not Sure",
                                                 "Don’t know/Not sure", # somehow different from the one above
                                                 "Refused",
                                                 "Don´t know/Not Sure",
                                                 "Don’t know / Not sure",
                                                 "Don’t know/Refused/Missing",
                                                 "Not asked or Missing"], np.nan)
```

Given that some attributes in the dataset are ordinal, I have decided to convert them to numeric to allow for comparisons between the values [10, p. 55].

```python
# converting nominal blood pressure attribute
BPHIGH_dict = {
    'No': 0,
    'Told borderline high or pre-hypertensive': 1,
    'Yes, but female told only during pregnancy': 2,
    'Yes': 3
}

for key in BPHIGH_dict:
    smoking_seniors_df.loc[(smoking_seniors_df.BPHIGH4 == key),'BPHIGH4'] = BPHIGH_dict[key]
```

```python
# converting nominal diabetes attribute
DIABET_dict = {
    'No': 0,
    'No, pre-diabetes or borderline diabetes': 1,
    'Yes, but female told only during pregnancy': 2,
    'Yes': 3
}

for key in DIABET_dict:
    smoking_seniors_df.loc[(smoking_seniors_df.DIABETE3 == key),'DIABETE3'] = DIABET_dict[key]
```

```python
# converting nominal general health attribute
GENHLTH_dict = {
    'Excellent': 5,
    'Very good': 4,
    'Good': 3,
    'Fair': 2,
    'Poor': 1,
}

for key in GENHLTH_dict:
    smoking_seniors_df.loc[(smoking_seniors_df.GENHLTH == key),'GENHLTH'] = GENHLTH_dict[key]
```

```python
# converting nominal BMI attribute
BMI_dict = {
    'Obese': 4,
    'Overweight': 3,
    'Normal Weight': 2,
    'Underweight': 1
}

for key in BMI_dict:
    smoking_seniors_df.loc[(smoking_seniors_df._BMI5CAT == key),'_BMI5CAT'] = BMI_dict[key]
```

```python
# selecting the rows representing older respondents
smoking_seniors_df = smoking_seniors_df[smoking_seniors_df['_IMPAGE'] == 'Age 65 or older'].reset_index(drop=True)

# dropping the age column, given that it now has a single value
smoking_seniors_df = smoking_seniors_df.drop('_IMPAGE', axis=1)

# selecting rows where respondents have either never smoked or quit smoking
smoking_seniors_df = smoking_seniors_df[(smoking_seniors_df['_SMOKER3'] == "Never smoked") | (smoking_seniors_df['_SMOKER3'] == "Former smoker")].reset_index(drop=True)

# dropping missing values
smoking_seniors_df = smoking_seniors_df.dropna().reset_index(drop=True)
```

```python
label_encoder = LabelEncoder()

# selecting columns to convert to numeric values
# all the below columns have only 2 values each, so will be converted to zeros and ones
cols_to_encode = ["PHYSHLTH",
                "MENTHLTH",
                "BLOODCHO",
                "CVDINFR4",
                "CVDCRHD4",
                "CVDSTRK3",
                "ASTHMA3",
                "CHCSCNCR",
                "CHCOCNCR",
                "CHCCOPD1",
                "HAVARTH3",
                "ADDEPEV2",
                "CHCKIDNY",
                "QLACTLM2",
                "USEEQUIP",
                "BLIND",
                "DECIDE",
                "DIFFWALK",
                "DIFFDRES",
                "DIFFALON",
                "EXERANY2",
                "_SMOKER3"]

for col in cols_to_encode:
    label_encoded = label_encoder.fit_transform(smoking_seniors_df[col])
    smoking_seniors_df[col] = label_encoded
```

```python
# viewing the first few rows of the dataset
smoking_seniors_df.head()
```

My question deals with predicting whether a respondent is a “former smoker” (true) or has “never smoked” (false). As I am choosing between two values, I felt that the Support Vector Machine method, specifically developed for binary classification [11] would be most suitable. I will build two models – one using a dataset of 25 health-related attributes, and then another using a reduced dataset of attributes selected using the SVM-based wrapper method.

In order to measure the accuracy of my classification models, I am going to use the Kappa statistic that expresses the result of the model as a value between random and perfect predictors [10, p. 181].

```python
# predicting smoking status
y = smoking_seniors_df['_SMOKER3']

# all other attributes in the dataset
X = smoking_seniors_df.drop('_SMOKER3', axis=1)

# splitting the data into a training and validation sets
train_X, val_X, train_y, val_y = train_test_split(X, y, test_size=0.2, random_state = 0) # 80% training and 20% test
```

```python
# Create a svm Classifier
clf = svm.SVC(kernel='linear') # Linear Kernel

# Train the model using the training sets
clf.fit(train_X, train_y)

# Predict the response for test dataset
y_preds = clf.predict(val_X)

# Model Accuracy: how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(val_y, y_preds))
print("Kappa score:",metrics.cohen_kappa_score(val_y, y_preds))
```

Selecting two attributes using the SVM-based wrapper method.

```python
# Sequential Forward Selection(sfs)
sfs = SFS(svm.SVC(kernel='linear'),
          k_features=2,
          forward=True,
          floating=False,
          scoring='accuracy',
          cv = 0)

sfs.fit(X, y)
sfs.k_feature_names_
```

Two attributes selected by the wrapper method are:

1. CVDINFR4 - Has a doctor, nurse, or other health professional ever told you that you had a heart attack, also called a myocardial infarction?


2. CHCCOPD1 - Has a doctor, nurse, or other health professional ever told you have (COPD) chronic obstructive pulmonary disease, emphysema or chronic bronchitis?

```python
# predicting smoking status
y = smoking_seniors_df["_SMOKER3"]

# two attributes representing past heart attack and chronic obstructive pulmonary disease
X2 = smoking_seniors_df[['CVDINFR4', 'CHCCOPD1']]

train_X, val_X, train_y, val_y = train_test_split(X2, y, test_size=0.2, random_state = 0) # 80% training and 20% test
```

```python
# Train the model using the training sets
clf.fit(train_X, train_y)

#Predict the response for test dataset
y_preds = clf.predict(val_X)

# Model Accuracy: how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(val_y, y_preds))
print("Kappa score:",metrics.cohen_kappa_score(val_y, y_preds))
```

Surprisingly, the results of the model trained on the dataset of 25 attributes and the model trained on 2 attributes are identical!

The Support Vector Machine indicates that past smoking does have an effect on the respondents’ heart and lung condition, which goes in line with research into tobacco harm [12]. The Kappa result of 0.2 indicates a slight relation [13], which may be explained by the fact that the former smokers may have quit some time ago. For instance, a risk of having a heart attack in halved 2 years after quitting smoking [14]. 

## Validity

The main threat to the validity of my analysis lies with the fact that the dataset used consists entirely of self-reported data, and is therefore threatened by self-reporting bias. Specifically, as my research questions concern sensitive topics such as tobacco use and diet, the data may be influenced by external bias due to social desirability or approval [15]. Furthermore, there is also a risk of measurement error bias – a respondent may have misunderstood an arguably convoluted question such as “During the past month … how many times per day, week, or month did you eat fruit?” [3], resulting in incorrect values entered.
